## Skala Diagram Terniary Basic

```python
# Import Library Tkinter dan math
from tkinter import *
import math

# Membuat variabel
## Posisi koordinat, panjang sisi, dan skala
posXa = 100                                      ## Inisiasi variabel posXa dengan nilai 100
posYa = 500                                      ## Inisiasi variabel posYa dengan nilai 500
sisi = 400                                       ## Inisiasi variabel sisi dengan nilai 400
skala = 40                                       ## Inisiasi variabel skala dengan nilai 40

# Statement untuk memasukan variabel A , B , dan C
print("Plot Data Lapangan: (contoh A=5, B=20, C=75)")                  ## Statement kata-kata awal sebagai judul dan contoh
A = float(input("Masukkan Plot Data Lapangan A (%):"))                 ## Stament kata-kata perintah masukan data A dan membuat variabel A
B = float(input("Masukkan Plot Data Lapangan B (%):"))                 ## Stament kata-kata perintah masukan data B dan membuat variabel B
C = 100 - A - B                                                        ## Membuat variabel C
print(f"Nilai dari A: {A} \nNilai dari B: {B} \nNilai dari C: {C}")    ## Statement kata-kata print nilai masing-masing variabel

# Pembuatan Skala Logaritmik
## Main Window
root = Tk()
root.title('Diagram Terniary')
root.geometry('700x700')

## Canvas
canvas = Canvas(root, height = 700, width = 700, bg = 'white')

# Membuat sisi-sisi Segitiga Terniary Utama
line = canvas.create_line(posXa, posYa, posXa + sisi, posYa, fill="blue")                                              ## Membuat garis B ke C
line = canvas.create_line(posXa + sisi, posYa, posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)), fill="red")    ## Membuat garis C ke A
line = canvas.create_line(posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)), posXa, posYa, fill="green")         ## Membuat garis B ke C

# Membuat statement untuk menampilkan teks nama Titik A, B, dan C
text = canvas.create_text(posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)) - 20, text="A")        ## Menamakan titik A
text = canvas.create_text(posXa - 15, posYa +15, text="B")                                                   ## Menamakan titik B
text = canvas.create_text(posXa + sisi + 15, posYa +15, text="C")                                            ## Menamakan titik C

# Membuat segitiga-segitia kecil di dalam Segitiga Terniary Utama
i = 0
while i < 10:
    ## Membuat garis garis bantu menuju titik C
    line = canvas.create_line(posXa + (i * skala), posYa, posXa + (sisi / 2) + (skala * i / 2), posYa - (sisi * math.sqrt(3) / 2) + (skala * i * math.sqrt(3) / 2), fill="green")
    ## Membuat garis garis bantu menuju titik B
    line = canvas.create_line(posXa + (i * skala), posYa, posXa + (sisi / 2) - (skala * (10 - i) / 2), posYa - (skala * i * math.sqrt(3) / 2), fill="red")
    ## Membuat garis garis bantu menuju titik A
    line = canvas.create_line(posXa + (skala * i /2 ), posYa - (skala * i * math.sqrt(3) / 2), posXa + (sisi - (i * skala / 2)), posYa - (skala * i * math.sqrt(3) / 2), fill="blue")
    i += 1

## Statement membuat data muncul di diagram teriniari dalam bentuk lingkaran
circle = canvas.create_oval((posXa + skala * (C + A / 2) / 10) - 5, (posYa - skala * A / 10 * math.sqrt(3) / 2) - 5, (posXa + skala * (C + A / 2) / 10) + 5, (posYa - skala * A / 10 * math.sqrt(3) / 2) + 5, fill='black')

# Membuat kolom nama dan nim
text = canvas.create_text(10,10,text="Klasifikasi Batuan Diagram Terniary Basic",anchor=W)     ## menamakan judul
text = canvas.create_text(10,25,text="Ponijo 190912837",anchor=W)                              ## nama dan NIM
text = canvas.create_text(10,55,text="Persentase Quartz (%):",anchor=W)                        ## "Persentase Quartz (%):"
text = canvas.create_text(10,70,text=A,anchor=W)                                               ## mengeluarkan variable A
text = canvas.create_text(10,85,text="Persentase Alkali Feldspar (%):",anchor=W)               ## "Persentase Alkali Feldspar (%):"
text = canvas.create_text(10,100,text=B,anchor=W)                                              ## mengeluarkan variable B
text = canvas.create_text(10,115,text="Persentase Plagioclase (%):",anchor=W)                  ## "Persentase Plagioclase (%):"
text = canvas.create_text(10,130,text=C,anchor=W)                                              ## mengeluarkan variable C


canvas.pack()
root.mainloop()
```

```python
# Import Library Tkinter dan math
from tkinter import *
import math

# Membuat variabel
## Posisi koordinat, panjang sisi, dan skala
posXa = 100                                      ## Inisiasi variabel posXa dengan nilai 100
posYa = 500                                      ## Inisiasi variabel posYa dengan nilai 500
sisi = 400                                       ## Inisiasi variabel sisi dengan nilai 400
skala = 40                                       ## Inisiasi variabel skala dengan nilai 40

# Statement untuk memasukan variabel A , B , dan C
print("Plot Data Lapangan: (contoh A=5, B=20, C=75)")                  ## Statement kata-kata awal sebagai judul dan contoh
A = float(input("Masukkan Plot Data Lapangan A (%):"))                 ## Stament kata-kata perintah masukan data A dan membuat variabel A
B = float(input("Masukkan Plot Data Lapangan B (%):"))                 ## Stament kata-kata perintah masukan data B dan membuat variabel B
C = 100 - A - B                                                        ## Membuat variabel C
print(f"Nilai dari A: {A} \nNilai dari B: {B} \nNilai dari C: {C}")    ## Statement kata-kata print nilai masing-masing variabel

# Pembuatan Skala Logaritmik
## Main Window
root = Tk()
root.title('Diagram Terniary')
root.geometry('700x700')

## Canvas
canvas = Canvas(root, height = 700, width = 700, bg = 'white')

# Membuat sisi-sisi Segitiga Terniary Utama
line = canvas.create_line(posXa, posYa, posXa + sisi, posYa, fill="blue")                                              ## Membuat garis B ke C
line = canvas.create_line(posXa + sisi, posYa, posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)), fill="red")    ## Membuat garis C ke A
line = canvas.create_line(posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)), posXa, posYa, fill="green")         ## Membuat garis B ke C

# Membuat statement untuk menampilkan teks nama Titik A, B, dan C
text = canvas.create_text(posXa + (sisi / 2), posYa - (sisi * (math.sqrt(3) / 2)) - 20, text="Quartz")             ## Menamakan titik A
text = canvas.create_text(posXa - 15, posYa + 15, text="Alkali Feldspar")                                          ## Menamakan titik B
text = canvas.create_text(posXa + sisi + 15, posYa + 15, text="Plagioclase")                                       ## Menamakan titik C

# Membuat segitiga-segitia kecil di dalam Segitiga Terniary Utama
i = 0
while i < 10:
    ## Membuat garis garis bantu menuju titik C
    line = canvas.create_line(posXa + (i * skala), posYa, posXa + (sisi / 2) + (skala * i / 2), posYa - (sisi * math.sqrt(3) / 2) + (skala * i * math.sqrt(3) / 2), fill="green")
    ## Membuat garis garis bantu menuju titik B
    line = canvas.create_line(posXa + (i * skala), posYa, posXa + (sisi / 2) - (skala * (10 - i) / 2), posYa - (skala * i * math.sqrt(3) / 2), fill="red")
    ## Membuat garis garis bantu menuju titik A
    line = canvas.create_line(posXa + (skala * i /2 ), posYa - (skala * i * math.sqrt(3) / 2), posXa + (sisi - (i * skala / 2)), posYa - (skala * i * math.sqrt(3) / 2), fill="blue")
    i += 1

## Statement membuat data muncul di diagram teriniari dalam bentuk lingkaran
circle = canvas.create_oval((posXa + skala * (C + A / 2) / 10) - 5, (posYa - skala * A / 10 * math.sqrt(3) / 2) - 5, (posXa + skala * (C + A / 2) / 10) + 5, (posYa - skala * A / 10 * math.sqrt(3) / 2) + 5, fill='black')

# Khusus Felsic Rock
## A> 5
nama_batuan = 0
data_salah = "Plot Data Lapangan Salah"
if A >= 90:
    nama_batuan = "Silexite"
elif A <= 5 and C <=10:
    nama_batuan = "Alkali Feldspar Syenite"
elif A <= 5 and C <=35:
    nama_batuan = "Syenite"
elif A <= 5 and C <=65:
    nama_batuan = "Monzonite"
elif A <= 5 and C <=90:
    nama_batuan = "Monzodiorite"
elif A <= 5 and C >=90:
    nama_batuan = "Diorite atau Gabbro atau Anorthosite"
## A > 20
elif A <= 20 and C <=10:
    nama_batuan = "Alkali Feldspar Quartz Syenite"
elif A <= 20 and C <=32:
    nama_batuan = "Quartz Syenite"
elif A <= 20 and C <=65:
    nama_batuan = "Quartz Monzonite"
elif A <= 20 and C <=90:
    nama_batuan = "Quartz Monzodiorite atau Quartz Monzogabbro"
elif A <= 20 and C >=90:
    nama_batuan = "Diorite atau Gabbro atau Anorthosite"
## A > 30
elif A > 20 and C <=10:
    nama_batuan = "Alkali Feldspar Granite"
elif A > 20 and C <=50:
    nama_batuan = "Granite"
elif A > 20 and B > 10:
    nama_batuan = "Granodiorite"
elif A > 20 and B < 10:
    nama_batuan = "Tonalite"
else:
    nama_batuan = data_salah

# Membuat kolom nama, nim, data plot, dan nama batuan
text = canvas.create_text(10,10,text="Klasifikasi Batuan Plutonik Felsik",anchor=W)     ## menamakan judul
text = canvas.create_text(10,25,text="Ponijo 190912837",anchor=W)                       ## nama dan NIM
text = canvas.create_text(10,55,text="Persentase Quartz (%):",anchor=W)                 ## "Persentase Quartz (%):"
text = canvas.create_text(10,70,text=A,anchor=W)                                        ## mengeluarkan variable A
text = canvas.create_text(10,85,text="Persentase Alkali Feldspar (%):",anchor=W)        ## "Persentase Alkali Feldspar (%):"
text = canvas.create_text(10,100,text=B,anchor=W)                                       ## mengeluarkan variable B
text = canvas.create_text(10,115,text="Persentase Plagioclase (%):",anchor=W)            ## "Persentase Plagioclase (%):"
text = canvas.create_text(10,130,text=C,anchor=W)                                       ## mengeluarkan variable C
text = canvas.create_text(10,145,text="Nama Batuan:",anchor=W)                          ## "Nama Batuan"
text = canvas.create_text(10,160,text=nama_batuan,anchor=W)                             ## mengeluarkan variable nama_batuan

    
canvas.pack()
root.mainloop()
```

# Codingan C++ (Jangan di run)

```python
#include <iostream>
#include <graphics.h>
#include <math.h>
#include <stdio.h> // digunakan untuk memanggil atau meanmbahkan file new
#include <cmath>
#include <string>
#include <fstream>

using namespace std;

int main()
{
    //Penentuan variabel dan inisiasi
    initwindow(600, 600, "Diagram Terniari");
    int posXa = 100; //inisiasi variabel posXa dengan nilai 100
    int posYa = 500; //inisiasi variabel posYa dengan nilai 500
    int sisi = 400; //inisiasi variabel sisi dengan nilai 400
    int skala = 40; //inisiasi variabel skala dengan nilai 40
    int A; //membuat variabel dengan nama A nanti dimasukan nilainya dengan cin A
    int B; //membuat variabel dengan nama B nanti dimasukan nilai dengan cin B
    int i; //membuat variabel dengan nama i sebagai variabel perulangan

    //Statement untuk memasukan variabel A , B , dan C
    cout << "Plot Data Lapangan: (contoh A=5, B=20, C=75" << endl; //Statement kata-kata awal sebagai judul dan contoh

    cout << "Masukkan Plot Data Lapangan A (%)"; //Stament kata-kata perintah masukan data A
    cin >> A; //statement untuk memasukan data ke dalam variabel A

    cout << "Masukkan Plot Data Lapangan B (%)"; //Statement kata-kata perintah masukan data B
    cin >> B; //statement untuk memasukan data ke dalam variabel B

    int C = 100 - A - B; //Membuat variabel C dengan asumsi A B C adalah 100


    //Satement untuk membuat sisi-sisi segitiga ternari
    line(posXa, posYa, posXa + sisi, posYa);
    line(posXa + sisi, posYa, posXa + (sisi / 2), posYa - (sisi * (sqrt(3) / 2)));
    line(posXa + (sisi / 2), posYa - (sisi * (sqrt(3) / 2)), posXa, posYa);

    //Statement untuk membuat segitiga-segitiga kecil di dalam segitiga terniari besar
    for (i=0; i<10; i++) //Membuat perintah perulangan for untuk segitiga segitiga kecil dengan variable i
    {
        setcolor(GREEN); //Membuat garis garis bantu menuju titik C
        line(posXa + (i * skala), posYa, posXa + (sisi / 2) + (skala * i / 2), posYa - (sisi * sqrt(3) / 2) + (skala * i * sqrt(3) / 2));

        setcolor(YELLOW); //Membuat garis garis bantu menuju titik B
        line(posXa + (i * skala), posYa, posXa + (sisi / 2) - (skala * (10 - i) / 2), posYa - (skala * i * sqrt(3) / 2));

        setcolor(CYAN); //Membuat garis garis bantu menuju titik A
        line (posXa + (skala * i /2 ), posYa - (skala * i * sqrt(3) / 2), posXa + (sisi - (i * skala / 2)), posYa - (skala * i * sqrt(3) / 2));
    }

    //Statement untuk menampilkan teks nama Titik A, B, dan C
    setcolor(WHITE);
    outtextxy(posXa - 10, posYa, "B"); //Menamakan titik B
    outtextxy(posXa + sisi + 5, posYa, "C"); //Menamakan titik C
    outtextxy(posXa + (sisi / 2) - 5, posYa - (sisi * (sqrt(3) / 2)) - 20, "A"); //Menamakan titik A

    //Statement memberikan nama Judul dan nama sendiri
    outtextxy(0,0,"Diagram Terniari"); //menamakan judul
    outtextxy(0,20, "Ponijo 190912837"); //nama dan NIM

    //statement membuat data muncul di diagram teriniari dalam bentuk lingkaran
    setcolor(WHITE);
    getmaxcolor();
    circle(posXa + skala * (C + A / 2) / 10, posYa - skala * A / 10 * sqrt(3) / 2, 8);


    getch();
    closegraph();
    return 0;
}
```
